```python
import matplotlib.pyplot as plt
import numpy as np
```

```python


binomial_data = np.random.binomial(n=10, p=0.5, size=1000)
plt.hist(binomial_data, bins=range(12), edgecolor='black', density=True)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.grid(True)
plt.show()

print("Mean of binomial simulation:", np.mean(binomial_data))
print("Expected mean (n * p = 5)")
```
