```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
df = pd.read_csv("online_Retail.csv", encoding='latin1')

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df['OrderDate'] = df['InvoiceDate'].dt.date
daily_orders = df.groupby('OrderDate')['Quantity'].sum()
high_order_days = daily_orders[daily_orders > 50]
n_days = len(daily_orders)
n_high = len(high_order_days)
```

```python
p_high = n_high / n_days
print(f"Total days: {n_days}")
print(f"High order days (>50 items): {n_high}")
print(f"Probability of high order day: {p_high:.2f}")
```

```python
x = np.arange(0, n_days + 1)
probabilities = binom.pmf(x, n_days, p_high)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, probabilities, marker='o', linestyle='-')
plt.title("Binomial Distribution of High Order Days")
plt.xlabel("Number of High Order Days")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```
