```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    "price": [300000, 450000, 500000, 600000, 750000, 800000, 950000],
    "sqft": [1500, 2000, 1800, 2500, 3000, 3200, 4000]
}
df = pd.DataFrame(data)
```

```python
summary_stats = df[["price", "sqft"]].describe()
print("=== House Price Summary Statistics ===")
print(summary_stats)
```

```python
df["price_per_sqft"] = df["price"] / df["sqft"]
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["sqft"], df["price"], color="blue", alpha=0.5)
plt.xlabel("Square Footage")
plt.ylabel("House Price ($)")
plt.title("House Price vs. Square Footage")
plt.grid(True)
plt.show()
```
